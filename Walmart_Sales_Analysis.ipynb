```python
#Splitting the date column into day,month and year
walmart_copy['Month'] = walmart_copy['Date'].str.split('/',expand = True)[0]
walmart_copy['Day'] = walmart_copy['Date'].str.split('/',expand = True)[1]
walmart_copy['Year'] = walmart_copy['Date'].str.split('/',expand = True)[2]
```

```python
#Dropping the 'Date' Column
walmart_copy.drop(['Date'],axis=1,inplace=True)
```

```python
#Checking the number of unique values of each column
walmart_copy.nunique()
```

```python
#Dropping the 'Invoice ID' column, as there are many unique values in that column.
walmart_copy.drop(['Invoice ID'],axis=1,inplace=True)
```

```python
# A. Analysing performance of sales and revenue at the city and branch level
# First, we create a new column 'Revenue' (Revenue = P * Q) so that it will easy for our further analysis
walmart_copy['Revenue'] = walmart_copy['Unit price'] * walmart_copy['Quantity']
```

```python
walmart_copy.head()
```

```python
#Analysing performance of sales and revenue at city level
a = sns.barplot(x = 'City',y = 'Quantity',data = walmart_copy,estimator = 'sum',errorbar = None)
a.bar_label(a.containers[0], fontsize=10)
```

The city 'Yangon' has the highest sum of sales amongst the three cities in observation for our analysis, followed by the cities 'Naypyitaw' and 'Mandalay' respectively.

```python
b = sns.barplot(x = 'City',y = 'Revenue',data = walmart_copy,estimator = 'sum',errorbar = None)
b.bar_label(b.containers[0], fontsize=10)
```

In the contrary, the city 'Naypyitaw' has the highest total revenue amongst the three cities in observation, followed by 'Yangon' and 'Mandalay' respectively.

Even though the total sum of sales is high in the city 'Yangon', the total revenue is quite low compared to the city 'Naypyitaw' because the unit price sold at 'Yangon' is quite less compared to 'Naypyitaw' (in my opinion).

```python
#Analysing performance of sales and revenue at branch level
c = sns.barplot(x = 'Branch',y = 'Quantity',data = walmart_copy,estimator = 'sum',errorbar = None)
c.bar_label(c.containers[0], fontsize=10)
```
